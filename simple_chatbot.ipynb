
# Simple chatbot using openai


```python
from openai import OpenAI ## import OpenAI
import os
from dotenv import load_dotenv ## dotenv is for environment variables
load_dotenv()

openai_api_key = os.getenv("OPENAI_API_KEY") ##OpenAI Key
cbot = OpenAI(api_key = openai_api_key) ##Key is added and stored as openai_client
print(f"openai key successfully added")
```

```python
def ai_simple_bot(your_question): 
    system_tone = "The results should be clearer with good behaviour and tone" ##instructing the bot about the way in which the 
    #reply should be. Basically tone and behaviour
    
    try:
        reply = cbot.chat.completions.create(model = 'gpt-4o-mini', messages = [{"role":"system","content":system_tone},
                                                                                {"role":"user","content":your_question}],
        temperature= 0.7)
        reply_msg = reply.choices[0].message.content
        return reply_msg

    except Exception as e:
        print(f"error,{e}")
        
```

```python
your_question = "who is lionel messi ?"
ai_simple_bot(your_question)
```
